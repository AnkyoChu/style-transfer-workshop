```python
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
MEAN_VALUES = np.array([123.68, 116.779, 103.939]).reshape((1,1,3))
```

```python
def preprocessing_function(img):
    MEAN_VALUES = np.array([123.68, 116.779, 103.939]).reshape((1,1,3))
    return img - MEAN_VALUES
```

```python
datagen = ImageDataGenerator(rotation_range=20,
                             width_shift_range=0.2,
                             height_shift_range=0.2,
                             vertical_flip=True,
                             preprocessing_function=preprocessing_function)
```

```python
generator = datagen.flow_from_directory('../COCO/',
                                        target_size=(256, 256),
                                        batch_size=32,
                                        class_mode=None)
```

```python
batches = generator.next()
```

```python
plt.imshow(np.clip(batches[0] + MEAN_VALUES, 0, 255).astype('uint8'))
```
